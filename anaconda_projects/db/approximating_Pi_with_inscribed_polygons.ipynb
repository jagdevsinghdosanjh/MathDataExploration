```python
import math
import numpy as np
import matplotlib.pyplot as plt

def polygon_pi_approx(n_sides):
    return n_sides * math.sin(math.radians(180 / n_sides))

sides = np.arange(3, 100, 2)
approximations = [polygon_pi_approx(n) for n in sides]

plt.plot(sides, approximations, marker="o", linestyle="-", label="Polygon Approximation")
plt.axhline(y=np.pi, color="red", linestyle="--", label="True π")
plt.xlabel("Number of Polygon Sides")
plt.ylabel("π Approximation")
plt.title("Approximating π with Inscribed Polygons")
plt.legend()
plt.show()
```
