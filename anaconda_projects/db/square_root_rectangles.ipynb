```python
import numpy as np
import matplotlib.pyplot as plt

sqrt_values = [np.sqrt(2), np.sqrt(3)]
fig, ax = plt.subplots(figsize=(6, 4))

# Draw rectangles
for i, sqrt_val in enumerate(sqrt_values):
    rect = plt.Rectangle((i, 0), 1, sqrt_val, edgecolor='green', facecolor='none', lw=2)
    ax.add_patch(rect)
    plt.text(i + 0.4, sqrt_val / 2, f"√{i+2}", fontsize=12)

plt.xlim(-0.5, len(sqrt_values) + 0.5)
plt.ylim(0, max(sqrt_values) + 0.5)
plt.title("Square Root Rectangles")
plt.show()
```
