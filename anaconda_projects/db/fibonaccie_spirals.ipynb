```python
import numpy as np
import matplotlib.pyplot as plt

# Fibonacci sequence
fib = [1, 1]
for _ in range(10):
    fib.append(fib[-1] + fib[-2])

# Golden ratio rectangles
x, y, w, h = 0, 0, 1, 1
rects = []

for i in range(len(fib)):
    rects.append([x, y, w, h])
    x += w
    w = h * 1.618  # Approximation of the golden ratio

# Plot rectangles
fig, ax = plt.subplots(figsize=(8, 6))
for r in rects:
    ax.add_patch(plt.Rectangle((r[0], r[1]), r[2], r[3], edgecolor='blue', facecolor='none', lw=2))

plt.xlim(0, x + w)
plt.ylim(0, max([r[3] for r in rects]) + 1)
plt.title("Golden Ratio Spiral Approximation")
plt.show()
```
