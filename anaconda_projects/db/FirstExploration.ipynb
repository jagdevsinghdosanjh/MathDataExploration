```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 1000)
y = np.sqrt(x)

plt.plot(x, y, label="Square Root Function", color='blue')
plt.xlabel("Number")
plt.ylabel("Square Root")
plt.title("Visualizing Square Roots of Non-Perfect Squares")
plt.legend()
plt.grid()
plt.show()
```
